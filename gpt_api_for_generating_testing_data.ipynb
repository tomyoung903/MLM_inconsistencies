```python
import requests
import os
import json

OPENAI_API_KEY = "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"

def ask_openai(message):
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {OPENAI_API_KEY}",
    }

    data = {
        "model": "gpt-3.5-turbo",
        "messages": [
            {"role": "user", "content": message}
        ]
    }

    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, data=json.dumps(data))

    # Ensure the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        json_response = response.json()

        # Get the assistant's message from the response
        assistant_message = json_response['choices'][0]['message']['content']

        return assistant_message
    else:
        print(f"Request failed with status code {response.status_code}")
```

```python
def verify_returned_words(returned_words):
    if not returned_words:
        return False
    # check no \n is in it
    if "\n" in returned_words:
        return False
    words = returned_words.split(" ")
    if len(words) > 5:
        return False
    if len(words) < 5:
        return False
    return True
```
